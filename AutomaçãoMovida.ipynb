```python
#-----------------------Iniciar---------------------------
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
import pandas as pd
from datetime import timedelta, datetime, date


# para rodar o chrome em 2º plano
#from selenium.webdriver.chrome.options import Options
#chrome_options = Options()
#chrome_options.headless = True 
#driver = webdriver.Chrome(options=chrome_options)

# abrir um navegador
driver = webdriver.Chrome()
# caso queira deixar na mesma pasta do seu código
# navegador = webdriver.Chrome("chromedriver.exe")
driver.maximize_window()


on = 1
i = 60
p = 1
k = 1
wait = WebDriverWait(driver, 20)
tabela = []

import pyautogui
import pyperclip

pyautogui.PAUSE = 1

#Definindo o caminho da pasta em que os arquivos serão salvos
mes_hoje = datetime.now().month
ano_atual = datetime.now().year
mes_ano = str(ano_atual) + "/" + str(mes_hoje)
```

```python
#------------------------Funções-------------------------

def clicknox():
    elementos = driver.find_elements(By.XPATH, '//*[@id="name"]')
    time.sleep(2)
    #xpath_close_modal = '//*[@id="mat-mdc-dialog-3"]/div/div/app-html-modal/div/div[1]/div[1]//*[@id="CloseModal"]/svg'
    #elementoX = driver.find_elements(By.XPATH, xpath_close_modal)
    
    # Verifica se a lista de elementos não está vazia antes de tentar acessar um índice específico
    if elementos:
        # Verifica se o elementoX foi encontrado
        #if elementoX:
            #try:                
                #time.sleep(2)  # Aguarde um curto período de tempo para garantir que o elemento esteja visível (opcional)

                # Realiza a ação de clique no botão CloseModal
                #close_modal_button.click()
                #print("Pop-up encontrado e fechado!")
            #except Exception as e:
                #print(f"Erro ao clicar no botão CloseModal: {e}")
        #else:
            #print("Botão CloseModal não encontrado.")
            
        pyautogui.press("esc")
    else:
        print("Pop Up não encontrado.")
        
def scroll_into_view(elementocar):
    driver.execute_script("arguments[0].scrollIntoView({ behavior: 'smooth', block: 'center' });", elementocar)

def clickcar(i):
    xpathcar = f"/html/body/app-root/div/div/showcase-all-cars/div/movida-show-result/div/div/div/div[6]/div[{i}]/a"
    
    try:
        # Aguardar até que o elemento seja visível
        elementocar = WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, xpathcar)))
        clicknox()
        # Role até o elemento usando a função scroll_into_view
        scroll_into_view(elementocar)
        time.sleep(3)
        # Clique no elemento
        elementocar.click()
        #print("Carro encontrado!")
    except Exception as e:
        on = 0
        print("Carro não encontrado!")
        
def click_km(p, prazo, k, car_description, car_name,tabela, mes_ano):
    on3 = 1
    
    while on3 == 1:
        xpathkm = f'//*[@id="container-step-1"]/div/div[2]/div[4]/ngx-slider/span[12]/span[{k}]/span'
        
        time.sleep(5)
        clicknox()
        elementokm = driver.find_elements(By.XPATH, xpathkm)
        
        if elementokm:
            try:
                elementokm[0].click()
                km = elementokm[0].text


                time.sleep(15)
                xpathvalor = '//*[@id="container-step-1"]/div/div[2]/div[5]/div[1]/p'
                elementovalor = driver.find_elements(By.XPATH, xpathvalor)
                valor = elementovalor[0].text
                
                xpathextra = '//*[@id="container-step-1"]/div/div[2]/div[5]/div[2]/p[2]/strong'
                elementoextra = driver.find_elements(By.XPATH, xpathextra)
                extra = elementoextra[0].text
                

                # Adicionando informações à tabela
                linha_tabela = [car_name, car_description, prazo, km, valor, extra, mes_ano]
                tabela.append(linha_tabela)

    #---------------------Print detalhes do carro-----------------------
                #print(f"Nome do carro: {car_name}")
                #print(f"Descrição do carro: {car_description}")
                #print(f"Prazo: {prazo}")
                #print(f"KM: {km}")
                #print(f"Valor: {valor}")
                print(linha_tabela)
                k = k + 1
                
            except Exception as e:
                clicknox()
        else:
            print("KM não encontrado!")
            on3 = 0

def click_prazo(p,k,car_description,car_name,tabela, mes_ano):
    on2 = 1
    while on2 == 1:
        xpathprazo = f'//*[@id="container-step-1"]/div/div[2]/div[2]/div[{p}]/span/span[1]'
        time.sleep(15)
        elementoprazo = driver.find_elements(By.XPATH, xpathprazo)
        
        if elementoprazo:
            try:
                elementoprazo[0].click()
                prazo = elementoprazo[0].text
                click_km(p, prazo, k, car_description, car_name, tabela, mes_ano)
                p = p + 1
            except Exception as e:
                clicknox()

            
        else:
            print("Prazo não encontrado!")
            on2 = 0
        
```

```python
#-----------------------Loop para todos os carros------------------------

while on == 1:
    driver.get("https://www.movidacarroporassinatura.com.br/assinatura/busca")
    time.sleep(15)
    clicknox()
    time.sleep(5)
    clickcar(i)

#-----------------------------Descrição do carro-------------------------------
    wait.until(EC.visibility_of_element_located((By.XPATH, '/html/body/app-root/div/div/car-detail-movida/movida-car-detail/div/div[1]/div[2]/movida-car-detail-signature/div[1]/div[1]/div[1]/div[1]/div/p')))
    time.sleep(5)
    clicknox()
    carro_description_ele = driver.find_elements(By.XPATH,
                                      '/html/body/app-root/div/div/car-detail-movida/movida-car-detail/div/div[1]/div[2]/movida-car-detail-signature/div[1]/div[1]/div[1]/div[1]/div/p')
    
    if carro_description_ele:
    # Obtém o texto do primeiro elemento na lista
        car_description = carro_description_ele[0].text
        
    else:
        print("Nenhum elemento encontrado.")
        
        
#----------------------------Nome do carro-----------------------------------------       
    clicknox()
    carro_name_ele = driver.find_elements(By.XPATH,
                                      '/html/body/app-root/div/div/car-detail-movida/movida-car-detail/div/div[1]/div[2]/movida-car-detail-signature/div[1]/div[1]/div[1]/div[1]/div/h3')
    if carro_name_ele:
    # Obtém o texto do primeiro elemento na lista
        car_name = carro_name_ele[0].text
        
        
    else:
        print("Nenhum elemento encontrado.")
    time.sleep(2)


#----------------------Clique botão de preços-----------------------------    
    clicknox()
    xpathplanobot = '/html/body/app-root/div/div/car-detail-movida/movida-car-detail/div/div[1]/div[2]/movida-car-detail-signature/div[1]/div[1]/div[2]/app-button/app-button-content'
    elementoplanobot = WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, xpathplanobot)))
    elementoplanobot.click()
    
    
    
#------------------------Prazos e KM----------------------- 
    click_prazo(p,k, car_description, car_name,tabela, mes_ano)
    i=i+1
    
    display(tabela)
    
    #DataFrame
    df = pd.DataFrame(tabela)

    #Nomeando as Colunas
    df = df[[0, 1,2, 3,4,5]].rename(columns= {0: 'Nome', 1: 'Descricao', 2:'Prazo', 3:'KM', 4:'Valor', 5: 'KM Excedente', 6:'Mes/Ano'})

    # Formatando Colunas
    df['Valor'] = df['Valor'].str.replace('/mês', '')

    df['Nome'] = df['Nome'].str.replace('\n', ' ')

    # Caminho/nome do arquivo Excel
    filename = '\Movida_' + str(ano_atual) + "_" + str(mes_hoje) + '.csv'
    caminho_csv = r'C:\Users\victor.rafael\Python Scripts\Movida_Valor'
    arquivo_csv = caminho_csv + filename

    # Salvar DataFrame no arquivo Excel
    df.to_csv(arquivo_csv, index=False, sep=";")
```
